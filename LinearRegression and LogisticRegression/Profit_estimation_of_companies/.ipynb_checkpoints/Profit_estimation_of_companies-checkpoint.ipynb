```python
#importing libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline
```

### Loading the dataset

```python
companies = pd.read_csv('1000_Companies.csv')
companies.head()
```

### Observing Correlation between Columns

```python
sns.heatmap(companies[['R&D Spend','Administration','Marketing Spend', 'Profit']].corr())
```

### Removing column due to high correlation

```python
companies.drop(['Marketing Spend'],axis=1,inplace=True)
```

```python
companies
```

### Encoding Data



```python
dummies=pd.get_dummies(companies.State)
```

```python
companies=pd.concat([companies,dummies],axis=1)
```

```python
companies
```

```python
companies.drop(['State'],axis=1,inplace=True)
companies
```

### Scaling

```python
from sklearn.preprocessing import MinMaxScaler
scale=MinMaxScaler()
companies[['R&D Spend','Administration']]=scale.fit_transform(companies[['R&D Spend','Administration']])
```

### Specifying X and y

```python
y=companies.iloc[:,2].values
y
```

```python
companies.drop(['Profit'],axis=1,inplace=True)
companies
```

```python
X=companies.iloc[:,:].values
X
```

### Splitting dataset into Train and Test

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=0)
```

### Fitting model to training set

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X_train,y_train)
```

### Predicting the Test dataset

```python
y_pred = lin_reg.predict(X_test)

y_pred
```

### Finding Coefficients and Intercepts

```python
# coefficient 
print(lin_reg.coef_)
```

```python
# intercepts
print(lin_reg.intercept_)
```

### Evaluating the model

```python
# calculating the  R squared error
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

### Predicting from random Values

```python
lin_reg.predict(([[1.0, 0.31665856662561165, False, False, False]]))
```
