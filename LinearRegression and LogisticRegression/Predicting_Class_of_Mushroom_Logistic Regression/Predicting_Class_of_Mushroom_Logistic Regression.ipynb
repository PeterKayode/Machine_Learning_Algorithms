### Importation of neccessary Libraries

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.metrics import r2_score
```

```python
mushroom_data = pd.read_csv("mushroom.csv")

mushroom_data
```

### Preprocessing the Data

```python
labelencoder = LabelEncoder()
for i in mushroom_data.columns:
    mushroom_data[i] = labelencoder.fit_transform(mushroom_data[i])

mushroom_data
```

### Splitting the data into Training and Testing data

```python
x = mushroom_data.drop("class", axis = 1)
y = mushroom_data["class"]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 42)
```

### Fitting the Logistic Model

```python
logistic_model = LogisticRegression(max_iter = 500)
logistic_model.fit(x_train, y_train)
```

### Predicting using the Model

```python
pred = logistic_model.predict(x_test)
pred
```

### Evaluating the model using Score

```python
logistic_model.score(x_test, y_test) 
```

```python
confusion_matrix(y_test, pred)
```

Out of the edible mushroom, we correctly measure 1201 as true and correctly measure 1114 as false, not edible.

We then label 68 as being edible that were actually poisonous. So we should decide not to use this model in the health domain
